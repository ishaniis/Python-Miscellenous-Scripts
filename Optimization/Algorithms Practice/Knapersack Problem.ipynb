```python
#Understanding

#Algorithm inspiration - Kie Codes

#Knapsack Problem 
#Step - 1 
#Genetic Representation of the Problem 

#Geonome is a combination of zeros and ones
#In our case , Genome is just the list of ones and zeros

#Where 0 represents the items is left out of the backpack
#Where 1 represents the items is taken into consideration to be put inside the backpack


from typing import List
from random import choices

#We'll generate the genome of 0 and 1 of a given length k = 

Genome = List[int]
Population = List[Genome]
Thing = namedtuple('Thing', ['name','value','weight'])

def generate_genome(length:int) -> Genome:
    return choices([0,1], k = length)

#k is the exact length of the list of the items to choose from the bag

#Now, we need a function to generate new solutions 
#i.e we need a population to generate new solutions

def generate_population(size: int, genome_length: int) -> Population:
    return [generate_genome(genome_length) for _ in range(size)]

#Now, We need a function which evaluates the generated solution for the given problem 
#Fitness Function

def fitness(genome: Genome, things: [Thing], weight_limit: int) -> int:
    if len(genome) != len(things):
        raise ValueError("Genome and things should be of same length")

        weight = 0
        value = 0
        
        for i, thing in enumerate(things):
            #We search for the solution along the given domain
            #We'll first check if its the part of the solution be checking the value of genome on various indices
            if genome[i] == 1:
                weight += thing.weight
                value += thing.value
                
                #If we dont hit the thresholds set for the weights then we'll return that value as a part of the solution
                #Else, if weight exceeds, we'll return zero
                
                if weight > weight_limit:
                    return 0
                #If it doesnt exceedsthe designated weight limit, we'll return the accumulated weight of the value
                
```
