```python
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
#Build a dataframe with your connections
df = pd.DataFrame({ 'from':['A' ,'A' ,'A','B' ,'C','C','D','D' ,'E','FINISH','G','H','H','I','I','J','J','K','L','M','N','O','P','Q','R','S', 
'T', 'U','V' ,'W','X','X', 'Y','START','START'], 'to':['O','P' ,'B','C','D','L','F','E','G','G','H','I','J','FINISH','H','I','T','I','F','K','M','N','Q','R','S','K','J','T','T','V','W', 'U' ,'X','A', 'Y']})
df
 
# Build your graph
G=nx.from_pandas_edgelist(df, 'from', 'to')
 
# Fruchterman Reingold
nx.draw(G, with_labels=True, node_size=150, node_color="skyblue", pos=nx.fruchterman_reingold_layout(G))
plt.title("fruchterman_reingold")
```

```python
# Spring
nx.draw(G,with_labels=True, center=('START','FINISH') , dim=2, resolution=0.35, equidistant=False, node_size=150, node_color="skyblue", pos=nx.spring_layout(G))
plt.title("spring")
```
